    <h4>Question 01 To Do List OOP Project(300 + 50):</h4>
    <p>You are tasked with creating a ToDo list manager using Object-Oriented Programming (OOP).
        Your program should consist of atleastthree classes: Task, Manager, and a Main function for
        running the application. Each class should be in a separate file to manage the project more
        effectively
    </p>
<p>Task Class:</p>
<ul>
<li>Each task must have the following attributes:</li>
<ul>
<li>A title (string).</li>
<li>A description(string).</li>
<li>A unique task ID, which should be a number but must be stored as a string (your program
should check thatthe ID is numeric).</li>
<li>Starttime, which is automatically recorded when a task is created.</li>
<li>End time, which is recorded when the task is marked as done.</li>
<li>Status: A boolean value thatis setto False when the task is created and setto True when the
task is marked as done.</li>
</ul>
</ul>
<p>Manager Class</p>
<ul>
<li>The manager should be able to:</li>
<ul>
<li>Add a task: Create and add a new task with a unique task ID.</li>
<li>Change task status to done: Mark the task as completed, update the status to True, and
record the end time.
</li>
<li>Display tasks: Show alltasks along with their details, such as task ID,title, description,
starttime, end time (if completed), and current status (done or not done).</li>
<li>Task summary: Display the number oftasks that are completed (status = True) and those
that are not completed (status = False).
</li>
</ul>
</ul>
<p>Main:</p>
<ul>
<li>The manager should be able to:</li>
<ul>
<li>The main function should interact with the Managerto allow users to perform actions
such as adding a task, marking a task as done, and viewing alltasks.</li>
</ul>
</ul>
<p style = "color:black;">Additional Requirements:</p>
<ul>
<li>Your program must be splitinto atleastthree files: one forthe Task class, one forthe Manager class, and
one forthe Main.</li>
<li>Make sure task IDs are unique</li>
<li>Ensure your program can handle and display the task details correctly.</li>
</ul>
</div>

```python
from datetime import datetime


class Task:
    to_do = {}

    def __init__(self, title: str, desc : str , id, status=False):
        self.id = id
        if not id.isnumeric():
            raise ValueError("Task ID must be numeric.")
        self.title = title
        self.desc = desc
        self.status = False
        self.stime = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    def add_task(self):
        if self.id not in Task.to_do:
            Task.to_do[self.id] = {
                'id': str(self.id),
                'title': self.title,
                'desc': self.desc,
                'start_time': self.stime,
                'status': self.status
            }
            print('Task added successfully')
        else:
            print('Task already exists')
        return Task.to_do
    def mark_done(self):
        if self.id in Task.to_do:
            Task.to_do[self.id]['status'] = True
            Task.to_do[self.id]['etime'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            print('Task marked as done')
        else:
            print('Task does not exist')
        return Task.to_do

    def __str__(self):
        status_str = "Done" if self.status else "Not Done"
        end_time_str = self.etime if self.etime else "N/A"
        return (f"Task ID: {self.task_id}, Title: {self.title}, Description: {self.desc}, "
                f"Start Time: {self.stime}, End Time: {end_time_str}, Status: {status_str}")






class Manager:
    def __init__(self):
        self.tasks = Task.to_do

    def add_task(self, title, desc, id):
        task = Task(title, desc, id)
        self.tasks = task.add_task()

    def mark_done(self, id):
        task = Task("", "", id)
        self.tasks = task.mark_done()

    def display_task(self):
        for task in self.tasks.values():
            print(task)
    
    def summery(self):
        done = 0
        not_done = 0
        for task in self.tasks.values():
            if task['status']:
                done += 1
            else:
                not_done += 1
        print(f"Total tasks: {len(self.tasks)}")
        print(f"Done: {done}")
        print(f"Not Done: {not_done}")

    def __str__(self):
        return f"Tasks: {self.tasks}"


def main():
    manager = Manager()
    while True:
        print("1. Add Task")
        print("2. Mark Task as Done")
        print("3. Display Tasks")
        print("4. Display Summary")
        print("5. Exit")
        print('---------------------------------')
        choice = input("Enter your choice: ")
        if choice == '1':
            title = input("Enter title: ")
            desc = input("Enter description: ")
            id = input("Enter ID: ")
            manager.add_task(title, desc, id)
        elif choice == '2':
            id = input("Enter ID of task to mark as done: ")
            manager.mark_done(id)
        elif choice == '3':
            manager.display_task()
        elif choice == '4':
            manager.summery()
        elif choice == '5':
            break
        else:
            print("Invalid choice")


if __name__ == '__main__':
    main()

```

<p>You are tasked with creating a system for managing electronic devices. Both Laptops and
Smartphones share some common features, butthey also have their own specific
characteristics. Use inheritance to avoid repeating code.<p>

</li>
</ul>
</ul>
<ul>
Laptop Class (Derived Class):
<ul>
<li>Inherits all attributes and methods from the Device class.
</li>
<li>Additional Attribute:</li>
<ul><li>RAM Size (integer,GB)</li></ul>
<li>Additional Method:</li>
<ul><li>Open Laptop: Prints a message indicating the laptop is opened.</li></ul>
</ul>
</ul>
<ul>
Smartphone Class (Derived Class):
<ul>

</li>

</ul>
</ul>

```python
class Device: 
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price

    def turn_on(self):
        print("Device is powered on")
    def turn_off(self):
        print("Device is powered off")
    def display(self):
        print(f'Brand : {self.brand}  , Model : {self.model} , Price : {self.price}')

class Laptop(Device):
    def __init__(self, brand, model, price,ram_size):
        super().__init__(brand, model, price)
        self.ram_s = ram_size

    def display(self):
        super().display()
        print(f' Ram : {self.ram_s} GB')

    def open_laptop(self):
        print('laprop is opened')

class SmartPhone(Device):
    def __init__(self, brand, model, price,camera_res):
        super().__init__(brand, model, price)   
        self.camera_resulation = camera_res
    
    def display(self):
        super().display()
        print(f'Camera Resulation: {self.camera_resulation} MP')
    def take_photo(self):
        print('Photo has been taken ')


def main():
    print(f'choose one option')
    print('1.is Device')
    print('2. is Laptop')
    print('3. is SmartPhone')
    print('4. Exit')

    x = input('Which is your choice?  ')

    if x == '1':
        b = input('give Brand name:')
        m  = input(' Give Model name: ')
        p = float(input('how much is Device Price: '))
        dev = Device(b,m,p)
        print('what do you want to do?')
        print('1.Display Device info')
        print('2.Turn off Device')
        print('3.Turn on Device')
        y = input('Which is your choice?  ')
        if y == '1':
            dev.display()
        if y == '2':
            dev.turn_off()
        if y == '3':
             dev.turn_on()
    if x == '2':
        b = input('give Brand name:')
        m  = input(' Give Model name: ')
        p = float(input('how much is Laptop Price: '))
        r = int(input('Give Size of  laptop Ram '))
        lap = Laptop(b,m,p,r)
        print('what do you want to do?')
        print('1.Display Laptop info')
        print('2.Turn off Device')
        print('3.Turn on Device')
        print('4.Open Laptop')
        y = input('Which is your choice?  ')
        if y == '1':
            lap.display()
        if y == '2':
            lap.turn_off()
        if y == '3':
            lap.turn_on()
        if y == '4':
            lap.open_laptop()
    if x == '3':
        b = input('give Brand name:')
        m  = input(' Give Model name: ')
        p = float(input('how much is Phone Price: '))
        c = int(input('Give Resulation of Phone Camera '))
        phone = SmartPhone(b,m,p,c)
        print('what do you want to do?')
        print('1.Display Laptop info')
        print('2.Turn off Device')
        print('3.Turn on Device')
        print('4. Take Photo')
        y = input('Which is your choice?  ')
        if y == '1':
            phone.display()
        if y == '2':
             phone.turn_off()
        if y == '3':
            phone.turn_on()
        if y == '4':
            phone.take_photo()
    if x =='4':
        print('exited')
        return
    



if __name__ == "__main__":
    main()

        

```
