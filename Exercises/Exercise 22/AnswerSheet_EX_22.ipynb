    <p>Create a digital store that manages two types of products: Phone and Laptop. Each product type has shared attributes, including an ID, name, price, and brand. Additionally, Laptop has specific attributes for memory and processor. In this project, you will define three classes: Phone, Laptop, and Store.</p> 
<p>Class Phone</p>
<li>Attributes:</li>
<li>id (str): The unique identifier of the product.</li>
<li>name (str): The name of the product.</li>
<li>price (float): The price of the product.</li>
<li>brand (str): The brand of the product.</li> 
<p>Methods:</p>
<li>display(): Display the product information.</li>

<p>Class Laptop</p>
<li>Attributes:</li>
<li>id (str): The unique identifier of the product.</li>
<li>name (str): The name of the product.</li>
<li>price (float): The price of the product.</li>
<li>brand (str): The brand of the product.</li> 
<li>memory (str): The memory of the laptop.</li>
<li>processor (str): The processor type of the laptop</li>
<p>Methods:</p>
<li>display(): Display the product information.</li>

<p>Class Store</p>
<li>Attributes:</li>
<li>phones: A collection to store phones.</li>
<li>laptops: A collection to store laptops.</li> 
<p>Methods:</p>
<li>add(): Add phone or laptop the phones collection.</li>
<li>emove(id): Remove phone or laptop collection using the ID</li>
<li>display(type): Display all phones or all laptops</li>
</div>

```python
class Phone: 
    def __init__(self,id_ =str , name =str ,price = float ,brand = str):
        self.id_ = id_
        self.name = name
        self.price = price 
        self.brand = brand
    
    def display(self):
        x = f"Phone INFO ==> Name : {self.name}  -- ID: {self.id_} -- Price: {self.price}  -- Brand: {self.brand} "
        print(x)

    
class Laptop:
    def __init__(self,id_ = str,name = str , price = float,brand = str,memory = str,processor = str):
        self.id_ = id_
        self.name = name
        self.price = price 
        self.brand = brand 
        self.memory = memory
        self.processor = processor

    def display(self):
        y = f"Laptop INFO ==> ID : {self.id_} ** Name : {self.name} ** Price : {self.price} ** Brand : {self.brand} ** Memory : {self.memory} ** Processor : {self.processor} "
        print(y)


class Store:
    def __init__(self):
        self.phones = []
        self.laptop = []
    
    def add(self,product):
        if isinstance(product,Phone):
            self.phones.append(product)
        elif isinstance(product,Laptop):
            self.laptop.append(product)
        else:
            print('Invalid Type of Product')

    def remove(self, id):
        for phone in self.phones:
            if phone.id == id:
                self.phones.remove(phone)
                print(f"Phone with {id} removed")
                return

        for laptop in self.laptop:
            if laptop.id == id: 
                self.laptop.remove(laptop)
                print(f"Laptop with {id} is been removed")
                return
        print(f"No Laptop Exists in this ID")

    def display(self, product_type):
        if product_type.lower() == 'phone':
            print("Displaying all phones:")
            for phone in self.phones:
                phone.display()
        elif product_type.lower() == 'laptop':
            print("Displaying all laptops:")
            for laptop in self.laptop:
                laptop.display()
        else:
            print("Invalid product type. Please specify 'phone' or 'laptop'.")



    
p1 = Phone('A14000', "Xiaomi",2000,"XO")
p2 = Phone('A14002', "Sumsong",3200,"BOB")
p3 = Phone('A14001', "Apple",2200,"James")
l1 = Laptop("L1000","Asus",1500,"Asus","16","xor")

s1 = Store()
s1.add(l1)
s1.add(p1)
s1.add(p2)
s1.add(p3)
s1.display('phone')
s1.display('laptop')        

    
```
